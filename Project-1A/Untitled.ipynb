```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import sys, platform, os
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import camb
from camb import model, initialpower
print('Using CAMB %s installed at %s'%(camb.__version__,os.path.dirname(camb.__file__)))
# make sure the version and path is what you expect
```

```python
# pars = camb.CAMBparams()
pars = camb.read_ini("./inifiles/params.ini")
```

```python
pars.Want_CMB
```

```python
pars
```

```python
pars.set_cosmology(H0=67.5, omch2=0.122, ombh2=0.022, mnu=0.06, omk=0, tau=0.06)
```

```python
pars.InitPower.set_params(As=2e-9, ns=0.965, r=0)
```

```python
pars.set_for_lmax(2500, lens_potential_accuracy=0)
```

```python
results = camb.get_results(pars)
```

```python
powers = results.get_cmb_power_spectra(pars, CMB_unit="muK")
```

```python
powers
```

```python
for name in powers: print(name)
```

```python
totCL = powers['total']
unlensedCL = powers['unlensed_scalar']
print(totCL.shape)
```

```python
ls = np.arange(totCL.shape[0])
fig, ax = plt.subplots(2,2,figsize=(12,12))
ax[0,0].plot(ls, totCL[:,0], color='k')
ax[0,0].plot(ls, unlensedCL[:,0], color='C2')
ax[0,0].set_title('TT')

ax[0,1].plot(ls[2:], 1-unlensedCL[2:,0]/totCL[2:,0])
ax[0,1].set_title("Fractional TT lensing")

ax[1,0].plot(ls, totCL[:,1], color='k')
ax[1,0].plot(ls, unlensedCL[:,1], color="C2")
ax[1,0].set_title("EE")

# ax[0,1].plot(ls, totCL[:,2], color='k')
# ax[0,1].plot(ls, unlensedCL[:,2], color='C2')
# ax[0,1].set_title('BB')

ax[1,1].plot(ls, totCL[:,3], color='k')
ax[1,1].plot(ls, unlensedCL[:,3], color='C2')
ax[1,1].set_title('TE')

for ax in ax.reshape(-1):
    ax.set_xlim([2,2500])
    ax.set_xlabel(r"$\ell$")
```
